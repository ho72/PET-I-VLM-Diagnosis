```python
import os
import urllib.parse
import sys

# -----------------------------------------------------------------
# â—ï¸â—ï¸â—ï¸ [1] ì‚¬ìš©ì ì„¤ì • â—ï¸â—ï¸â—ï¸
#
# Label Studioì—ì„œ exportí•œ 'labels' í´ë”ì˜ ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš”.
# (ì˜ˆ: r"C:\Users\MyUser\Desktop\yolo_dataset\labels")
#
# -----------------------------------------------------------------
labels_directory = r"./yolo_data/labels"
# -----------------------------------------------------------------


def rename_label_files(directory):
    """
    Label Studio export íŒŒì¼ëª…(ì˜ˆ: hash-url_encoded_name.txt)ì„
    ì›ë³¸ íŒŒì¼ëª…(ì˜ˆ: original_name.txt)ìœ¼ë¡œ ë³€ê²½í•©ë‹ˆë‹¤.
    """
    
    if not os.path.isdir(directory):
        print(f"âŒ ì˜¤ë¥˜: '{directory}' í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        print("ìŠ¤í¬ë¦½íŠ¸ ìƒë‹¨ì˜ 'labels_directory' ë³€ìˆ˜ ê²½ë¡œë¥¼ ì˜¬ë°”ë¥´ê²Œ ìˆ˜ì •í•´ì£¼ì„¸ìš”.")
        return

    print(f"ğŸ“ ì‘ì—… ëŒ€ìƒ í´ë”: {directory}\n")
    
    renamed_count = 0
    skipped_count = 0

    try:
        # ë””ë ‰í„°ë¦¬ì˜ ëª¨ë“  íŒŒì¼ ëª©ë¡ì„ ê°€ì ¸ì˜µë‹ˆë‹¤.
        for filename in os.listdir(directory):
            
            # .txt íŒŒì¼ì´ ì•„ë‹ˆë©´ ê±´ë„ˆëœë‹ˆë‹¤.
            if not filename.endswith(".txt"):
                continue

            # 1. íŒŒì¼ ê²½ë¡œë¥¼ ë§Œë“­ë‹ˆë‹¤.
            old_filepath = os.path.join(directory, filename)

            # 2. ì²« ë²ˆì§¸ í•˜ì´í”ˆ(-)ì„ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•©ë‹ˆë‹¤.
            parts = filename.split('-', 1)

            # 3. í•˜ì´í”ˆì´ ìˆê³  (ì¦‰, 2ì¡°ê°ìœ¼ë¡œ ë‚˜ë‰˜ì—ˆê³ )
            if len(parts) == 2:
                try:
                    # 4. í•˜ì´í”ˆ ë’¤ì˜ ë¶€ë¶„ì„ URL ë””ì½”ë”©í•©ë‹ˆë‹¤. (ì˜ˆ: valid_ë°±ë‚´ì¥_90.txt)
                    new_filename = urllib.parse.unquote(parts[1])
                    
                    # 5. ìƒˆ íŒŒì¼ ê²½ë¡œë¥¼ ë§Œë“­ë‹ˆë‹¤.
                    new_filepath = os.path.join(directory, new_filename)

                    # 6. (ì•ˆì „ì¥ì¹˜) í˜¹ì‹œ ë³€ê²½ë  ì´ë¦„ì´ ì´ë¯¸ ì¡´ì¬í•˜ë©´ ê±´ë„ˆëœë‹ˆë‹¤.
                    if os.path.exists(new_filepath):
                         print(f"âš ï¸  ê±´ë„ˆëœ€: '{new_filename}' íŒŒì¼ì´ ì´ë¯¸ ì¡´ì¬í•©ë‹ˆë‹¤. (ì›ë³¸: {filename})")
                         skipped_count += 1
                         continue

                    # 7. íŒŒì¼ ì´ë¦„ì„ ë³€ê²½í•©ë‹ˆë‹¤.
                    os.rename(old_filepath, new_filepath)
                    print(f"  âœ… ë³€ê²½: {filename}  ->  {new_filename}")
                    renamed_count += 1

                except Exception as e:
                    print(f"  âŒ ì˜¤ë¥˜ (ì²˜ë¦¬ ì¤‘): {filename} -> {e}")
                    skipped_count += 1
            else:
                # í•˜ì´í”ˆì´ ì—†ëŠ” íŒŒì¼ (ì´ë¯¸ ì´ë¦„ì´ ì˜¬ë°”ë¥´ê±°ë‚˜ ë‹¤ë¥¸ íŒŒì¼)
                print(f"  â„¹ï¸  ê±´ë„ˆëœ€: {filename} (í•˜ì´í”ˆ ì ‘ë‘ì‚¬ ì—†ìŒ)")
                skipped_count += 1
                
    except FileNotFoundError:
        print(f"âŒ ì˜¤ë¥˜: '{directory}' í´ë”ë¥¼ ì°¾ëŠ” ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")
        return
    except Exception as e:
        print(f"âŒ ì•Œ ìˆ˜ ì—†ëŠ” ì˜¤ë¥˜ ë°œìƒ: {e}")
        return

    print("\n--- âœ¨ ì‘ì—… ì™„ë£Œ âœ¨ ---")
    print(f"ì´ ë³€ê²½ëœ íŒŒì¼: {renamed_count}ê°œ")
    print(f"ì´ ê±´ë„ˆë›´ íŒŒì¼: {skipped_count}ê°œ")


if __name__ == "__main__":
    # ê²½ë¡œê°€ ê¸°ë³¸ê°’ì¸ì§€ í™•ì¸
    if labels_directory == r"C:\ì—¬ê¸°ì—\labels\í´ë”ì˜\ì „ì²´\ê²½ë¡œë¥¼\ì…ë ¥í•˜ì„¸ìš”":
        print("="*60)
        print("â—ï¸ ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•˜ê¸° ì „ì— 'labels_directory' ë³€ìˆ˜ì˜ ê°’ì„")
        print("   ì‹¤ì œ 'labels' í´ë” ê²½ë¡œë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”.")
        print("="*60)
    else:
        rename_label_files(labels_directory)
```

### yolo

```python
import os
import random
import shutil

# -----------------------------------------------------------------
# â—ï¸ [1] ì‚¬ìš©ì ì„¤ì •: ì›ë³¸ ë°ì´í„°ê°€ ìˆëŠ” í´ë” ê²½ë¡œ
# -----------------------------------------------------------------
# 'images'ì™€ 'labels' í´ë”ê°€ ë“¤ì–´ìˆëŠ” ìµœìƒìœ„ í´ë” ê²½ë¡œ
dataset_root = r"./yolo_data"

# (ì˜µì…˜) í•™ìŠµìš© ë°ì´í„° ë¹„ìœ¨ (0.8 = 80%)
train_ratio = 0.8
# -----------------------------------------------------------------

# 1. ì›ë³¸ í´ë” ê²½ë¡œ
images_src_dir = os.path.join(dataset_root, "images")
labels_src_dir = os.path.join(dataset_root, "labels")

# 2. ìƒˆë¡œ ë§Œë“¤ í´ë” ê²½ë¡œ ì •ì˜
train_img_dir = os.path.join(dataset_root, "train", "images")
train_lbl_dir = os.path.join(dataset_root, "train", "labels")
valid_img_dir = os.path.join(dataset_root, "valid", "images")
valid_lbl_dir = os.path.join(dataset_root, "valid", "labels")

# 3. í´ë” ìƒì„±
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(valid_img_dir, exist_ok=True)
os.makedirs(valid_lbl_dir, exist_ok=True)

# 4. ì´ë¯¸ì§€ íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
try:
    image_files = [f for f in os.listdir(images_src_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
    if not image_files:
        print(f"âŒ ì˜¤ë¥˜: '{images_src_dir}' í´ë”ì— ì´ë¯¸ì§€ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        exit()
        
    random.shuffle(image_files) # íŒŒì¼ ì„ê¸°

    # 5. í•™ìŠµìš©/ê²€ì¦ìš© ë¶„ë¦¬
    split_point = int(len(image_files) * train_ratio)
    train_files = image_files[:split_point]
    valid_files = image_files[split_point:]

    print(f"ì´ ì´ë¯¸ì§€: {len(image_files)}ê°œ")
    print(f"í•™ìŠµìš©(train): {len(train_files)}ê°œ")
    print(f"ê²€ì¦ìš©(valid): {len(valid_files)}ê°œ")

    # 6. íŒŒì¼ ì´ë™ í•¨ìˆ˜
    def move_files(file_list, img_dest, lbl_dest):
        moved_count = 0
        for img_file in file_list:
            # íŒŒì¼ëª… (í™•ì¥ì ì œì™¸)
            base_filename = os.path.splitext(img_file)[0]
            label_file = base_filename + ".txt"
            
            # ì›ë³¸ ê²½ë¡œ
            img_src_path = os.path.join(images_src_dir, img_file)
            lbl_src_path = os.path.join(labels_src_dir, label_file)

            # ëŒ€ìƒ ê²½ë¡œ
            img_dest_path = os.path.join(img_dest, img_file)
            lbl_dest_path = os.path.join(lbl_dest, label_file)

            # íŒŒì¼ ì´ë™
            if os.path.exists(lbl_src_path):
                shutil.move(img_src_path, img_dest_path)
                shutil.move(lbl_src_path, lbl_dest_path)
                moved_count += 1
            else:
                print(f"âš ï¸  ê²½ê³ : '{img_file}'ì˜ ì§ê¿ íŒŒì¼ '{label_file}'ì„(ë¥¼) ì°¾ì„ ìˆ˜ ì—†ì–´ ê±´ë„ˆëœë‹ˆë‹¤.")
        return moved_count

    # 7. ì‹¤ì œ íŒŒì¼ ì´ë™ ì‹¤í–‰
    print("\n[í•™ìŠµìš© ë°ì´í„° ì´ë™ ì¤‘...]")
    move_files(train_files, train_img_dir, train_lbl_dir)
    
    print("\n[ê²€ì¦ìš© ë°ì´í„° ì´ë™ ì¤‘...]")
    move_files(valid_files, valid_img_dir, valid_lbl_dir)

    print("\n--- âœ¨ ë°ì´í„° ë¶„ë¦¬ ì™„ë£Œ! ---")
    print(f"'images'ì™€ 'labels' ì›ë³¸ í´ë”ëŠ” ì´ì œ ë¹„ì–´ìˆì–´ì•¼ í•©ë‹ˆë‹¤.")

except FileNotFoundError:
    print(f"âŒ ì˜¤ë¥˜: '{images_src_dir}' ë˜ëŠ” '{labels_src_dir}' í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    print("ìŠ¤í¬ë¦½íŠ¸ ìƒë‹¨ì˜ 'dataset_root' ê²½ë¡œë¥¼ ì˜¬ë°”ë¥´ê²Œ ìˆ˜ì •í•´ì£¼ì„¸ìš”.")
```
